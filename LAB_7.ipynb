```python
# -------------------------
# Knowledge Base Clauses
# -------------------------
clauses = [
    {"~Food(x)", "Likes(John,x)"},   # John likes all food
    {"Food(Apple)"},                 # Apple is food
    {"Food(Vegetables)"},            # Vegetables are food
    {"~Eats(x,y)", "~Alive(x)", "Food(y)"},  # Eats + Alive -> Food
    {"Eats(Anil,Peanuts)"},          # Anil eats peanuts
    {"Alive(Anil)"},                 # Anil is alive
    {"~Alive(x)", "~Killed(x)"},     # Alive -> not killed
    {"Killed(x)", "Alive(x)"},       # not killed -> alive
    {"~Likes(John,Peanuts)"}         # NEGATED QUERY
]


# -------------------------
# Unification Helpers
# -------------------------

def unify(x, y, subs):
    if subs is None:
        return None
    elif x == y:
        return subs
    elif isinstance(x, str) and x.islower():
        return unify_var(x, y, subs)
    elif isinstance(y, str) and y.islower():
        return unify_var(y, x, subs)
    return subs


def unify_var(var, x, subs):
    if var in subs:
        return unify(subs[var], x, subs)
    elif x in subs:
        return unify(var, subs[x], subs)
    else:
        subs[var] = x
        return subs


# -------------------------
# Resolution Rule
# -------------------------

def resolve(c1, c2):
    """
    Try to resolve two clauses.
    Example: {A, ~B} & {B, C} → {A, C}
    """
    for lit1 in c1:
        for lit2 in c2:

            # Case: ~P and P resolve
            if lit1.startswith("~") and lit1[1:] == lit2:
                return (c1 - {lit1}) | (c2 - {lit2})

            # Case: P and ~P resolve
            if lit2.startswith("~") and lit2[1:] == lit1:
                return (c1 - {lit1}) | (c2 - {lit2})

    return None


# -------------------------
# FOL Resolution Algorithm
# -------------------------

def fol_resolution(clauses):
    clauses = [frozenset(c) for c in clauses]
    new = set()

    while True:
        n = len(clauses)

        for i in range(n):
            for j in range(i + 1, n):

                resolvent = resolve(clauses[i], clauses[j])

                # Empty clause → contradiction proved
                if resolvent == set():
                    return True

                if resolvent is not None:
                    new.add(frozenset(resolvent))

        # No progress → cannot prove
        if new.issubset(set(clauses)):
            return False

        # Add new clauses
        for c in new:
            if c not in clauses:
                clauses.append(c)


# -------------------------
# Run Resolution Proof
# -------------------------

result = fol_resolution(clauses)
print("Conclusion:", result)
```
