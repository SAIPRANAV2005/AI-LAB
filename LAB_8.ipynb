```python
INF = 999999
pruned_nodes = []      # to track pruned branches
path = []              # best path from root

def alpha_beta(node, depth, alpha, beta, is_maximizing, tree, current_path):
    global pruned_nodes, path

    # If leaf node:
    if node not in tree:      # node is a leaf value
        return node

    # -------------------- MAXIMIZING PLAYER --------------------
    if is_maximizing:
        value = -INF
        best_child = None

        for child in tree[node]:
            new_path = current_path + [child]

            child_val = alpha_beta(child, depth - 1, alpha, beta, False, tree, new_path)

            if child_val > value:
                value = child_val
                best_child = new_path

            alpha = max(alpha, value)

            # ---- Pruning ----
            if alpha >= beta:
                index = tree[node].index(child)
                remaining = tree[node][index + 1:]
                pruned_nodes.extend(remaining)
                break

        if best_child:
            path.clear()
            path.extend(best_child)

        return value

    # -------------------- MINIMIZING PLAYER --------------------
    else:
        value = INF

        for child in tree[node]:
            new_path = current_path + [child]

            child_val = alpha_beta(child, depth - 1, alpha, beta, True, tree, new_path)

            if child_val < value:
                value = child_val

            beta = min(beta, value)

            # ---- Pruning ----
            if alpha >= beta:
                index = tree[node].index(child)
                remaining = tree[node][index + 1:]
                pruned_nodes.extend(remaining)
                break

        return value


# -------------------- Example Game Tree --------------------
tree = {
    'A': ['B', 'C', 'D'],   # MAX
    'B': [3, 5, 6],         # MIN
    'C': [9, 1, 2],         # MIN
    'D': [0, -1, 4]         # MIN
}

root = 'A'
root_value = alpha_beta(root, depth=10, alpha=-INF, beta=INF,
                        is_maximizing=True, tree=tree, current_path=[root])

print("Root Value:", root_value)
print("Best Path:", " → ".join(map(str, path)))
print("Pruned Nodes:", pruned_nodes)
```
